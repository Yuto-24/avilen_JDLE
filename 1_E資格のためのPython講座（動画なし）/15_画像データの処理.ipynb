```python
# numpyで画像処理を行う準備
import numpy as np
from PIL import Image
from matplotlib import pylab as plt
```

## 1.画像の読み込みと表示  
まず画像をndarrayとして読み込む方法とそれを表示する方法を見て行きましょう．Pillowを用いて画像を読み込み，それをndarrayとすることで画像を読み込むことができます．表示にはmatplotlibを用います．  
【例題】画像の読み込みと表示を行います．以下のコードを読んで実行してみましょう．

```python
# 画像の読み込み
image = np.array(Image.open('Lenna.jpg'))

# 画像の表示
plt.imshow(image)
plt.show
```

- Lenna.jpgが表示されたことを確認しましょう．  
- ファイル名Lenna.jpgが文字列になっていることを確認しましょう．  
- 別の場所に画像がある場合はパスを渡す必要があります．

## 2.画像データの構造  
それでは読み込んだ画像データの構造を見てみましょう．

```python
print(image.shape)
```

- 読み込まれた画像データは三次元であり，二次元がxy平面，一次元が色となっています．色はRGBの順番で格納され，赤:R,緑G,青:B,それぞれの値によって色を表現しています．  
- 読み込まれた画像データは縦512,横512であり，それぞれの座標において，一つのピクセルが3つの要素RGBを持っていることになります．

次にピクセル値を見てみます．画像データの各ピクセルの各色は0～255の値を持っており，この値が大きければその色は明るく，小さければ暗いことになります．3色すべてが0であれば真っ黒の画像となり，255であれば真っ白な画像となります．  
【例題】読み込んだ画像の100行100列目のピクセル値を見てみます．以下のコードを読んで実行してみましょう．

```python
# 100行100列目のピクセル値を表示
print(image[100, 100])
```

- [ 181 74 84 ]となっていれば成功です．

読み込んだ画像はndarrayなので，image[:, :, 0]などとすることで各ピクセルにおける指定した色の値を得ることができます．  

【問題】上述した色の値を取得する手法を使って，読み込んだ画像をグレースケール画像にしてみましょう．グレースケール画像はR,G,Bそれぞれのチャンネルのピクセルの平均を取ることで得られます．

```python
# grayにimageの値をコピー
gray = image.copy()
# grayに格納された画像のそれぞれの色(チャンネルごと）について，平均をとる．
gray[:, :, 0] = np.mean(image, axis=2)
gray[:, :, 1] = np.mean(image, axis=2)
gray[:, :, 2] = np.mean(image, axis=2)

# 表示
plt.imshow(gray)
plt.show
```

- グレースケールの濃淡画像が表示されれば正解です．

## 1. 配列の要素のシャッフル   
配列の要素をシャッフルする場合，numpy.random.permutation()を使うことができます．引数にはintかリスト，配列を与えることができます．この関数は配列をシャッフルし，シャッフルした配列のコピーを生成します．引数にint型のxを渡した場合，np.arange(x)の要素をシャッフルしたものを生成し，リストまたは配列を渡した場合は要素をシャッフルします．  
【例題】それでは配列の要素をシャッフルしてみます．以下のコードを読んで実行してみましょう．

```python
# 配列の生成
numpy_1 = np.arange(9).reshape(3, 3)
print('配列')
print(numpy_1)

# 配列の要素をシャッフル
numpy_2 = np.random.permutation(numpy_1)

print('numpy_1')
print(numpy_1)
print('numpy_2')
print(numpy_2)
```

- numpy_1がシャッフルする前の配列で，numpy_2がシャッフルした後の配列です．  
- numpy_2の要素がシャッフルされていれば成功です．numpy_1がシャッフルされてnumpy_2が生成されても，numpy_1は上書きされずそのままであることを確認しましょう．また，要素のシャッフルの軸は行方向です．

【問題】以下の配列をnumpy.random.permutation()を用いて要素をシャッフルしてみましょう．

```python
# 配列の生成
numpy_1 = np.arange(9, 0, -1).reshape(3, 3)
print(numpy_1)
print('--------------------')

# 配列の要素のシャッフル
numpy_2 = np.random.permutation(numpy_1)
print(numpy_2)
```

- 配列の要素がランダムに並べ替えられていれば成功です．

#### ミニバッチ処理で用いる手法  
次にミニバッチ処理の基本となる，シャッフルされたデータを指定された個数ずつ選ぶ方法を解説します．  
10個ずつ選ぶやり方を考えます．まずnumpy.random.permutation()を用いてデータをシャッフルし，次にnp.arange()でシャッフルされたデータの配列に対して10個ずつの等差数列を生成します．等差数列を用いてfor文を回し，for文の中でデータを取り出します．  
以下のコードを読んで実行してみましょう．

```python
# データとバッチサイズの定義
data = np.arange(50)
print('データ')
print(data)
batch_size = 10
cnt = 0

# データのインデックスをシャッフル
perm = np.random.permutation(len(data))

# 10個ずつデータを取り出す
for idx in np.arange(0, len(perm), batch_size): # 等差10の等差数列を用いる
    cnt = cnt + 1
    x = data[perm[idx:idx + batch_size]] # データの取り出し
    print("ミニバッチ" + str(cnt))
    print(x)
```

## 2. 配列から要素をランダムに抽出する  
一次元の配列から要素をランダムに抽出したい場合は，numpy.random.choice()を使うことができます．必要に応じて出現率を指定することも可能です．  
numpy.random.choice(a, size, replace, p)となっており，引数の役割はそれぞれ a:一次元の配列またはint, size:配列から抽出する要素の個数, replace:ブール値で指定し，Trueなら要素の重複あり, p:与えられた配列に対する出現率をリストで指定 となっています．  
【例題】配列から要素をランダムに抽出します．以下のコードを読んで実行してみましょう．

```python
# 配列の生成
numpy_1 = np.arange(3)
print('並べ替える前', numpy_1)

# 要素の抽出
numpy_2 = np.random.choice(numpy_1, 3, replace = True, p=[0.2, 0.2, 0.6])
print('並べ替えた後', numpy_2)
```

- 配列が並べ替えられていれば成功です．
- 引数で指定した数だけ要素が抽出されていることを確認しましょう．

【問題】以下の配列の要素を，重複なしで5つ抽出してみましょう．出現率はそれぞれの要素に対して一定とします．

```python
# 配列の生成
numpy_1 = np.arange(10)

# 要素の抽出
numpy_2 = np.random.choice(numpy_1, 5, False)
print(numpy_2)
```

- 配列の要素を重複なしで5個抽出できれば正解です．

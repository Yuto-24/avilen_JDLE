```python
# numpyとmatplotlibのインポート
import numpy as np
import matplotlib.pyplot as plt
```

## 1.一様分布  
numpy.randomモジュールを用いることで乱数に関する関数を使うことができます.  
numpy.random.rand()は一様分布の乱数を返します(0.0以上1.0未満の乱数)．引数には生成する配列のサイズを整数で指定することができます．また，任意の範囲の乱数を取得したい場合は，[a, b)の場合，(b - a) * np.random.rand() + a とすることで取得することができます．  
【例題】一様分布の乱数を生成してみます．以下のコードを読んで実行してみましょう．

```python
# 乱数を一つ取得
rand = np.random.rand()
print("1つの配列")
print(rand)

# サイズが4の配列を取得
rand = np.random.rand(4)
print("\n4の配列 " + str(rand))

# 3 x 3の配列の乱数を取得
rand = np.random.rand(3, 3)
print("\n3 x 3の配列\n" +str(rand))
```

- それぞれ引数に指定したサイズに応じた形状の一様分布の乱数が取得できていることを確認しましょう．  

【問題】50以上100未満の乱数を生成してみましょう．

```python
rand = 50 * np.random.rand() + 50
print(rand)
```

- 生成した乱数が50-100の間に収まっていれば正解です．

## 2.正規分布  
任意の平均，標準偏差の正規分布に従う乱数を取得するにはnumpy.random.normal()を用います．引数には平均loc,標準偏差scale,サイズ（タプル)sizeで指定することができます．  
【例題】正規分布に従う乱数を生成します．以下のコードを読んで実行してみましょう．

```python
# 乱数の生成
rand = np.random.normal(loc = 170, scale = 6, size = 100000)
# ヒストグラムの表示
plt.hist(rand, bins = 100)
```

- ヒストグラムの形状から，生成された乱数が正規分布に従うことを確認しましょう．

【問題】平均314,標準偏差86.9とした正規分布に従う乱数を100,000個生成してみましょう．

```python
# 乱数の生成
rand = np.random.normal(loc = 314, scale = 86.9, size = 100000)
# 表示
plt.hist(rand, bins = 100)
```
